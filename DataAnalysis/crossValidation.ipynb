## Environment comparisons

```python
from DataAnalysis.Analysis import plotAlgorithmEpisodeAverageReward
import pickle

filename = 'ExperimentFramework/Experiments/Results/ESCrossValidation.pkl'
f = open(filename, 'rb')
stepSize = 8
data = pickle.load(f)
algorithms = list(data["algorithms"].keys())
numsAgents = list(data["algorithms"][algorithms[0]].keys())[1:]
environments = list(data["algorithms"][algorithms[0]][numsAgents[0]].keys())
print(algorithms)
print(numsAgents)
print(environments)
for numAgents in numsAgents:
    for environment in environments:
        for i in range(0,len(algorithms), stepSize):
            plotAlgorithmEpisodeAverageReward(data, algorithms[i:i+stepSize], numAgents, environment)
f.close()
```
